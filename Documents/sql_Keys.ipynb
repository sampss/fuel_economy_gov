## SQL Keywords
**Data Definition Language (DDL)**
- **CREATE**: Used to create a new table, view, or other database object.
- **ALTER**: Modifies an existing database object, such as a table.
- **DROP**: Deletes an existing database object.
- **RENAME**: Changes the name of a database object.
- **TRUNCATE**: Removes all records from a table, but not the table itself.

**Data Manipulation Language (DML)**
- **SELECT**: Retrieves data from a database.
- **INSERT**: Adds new rows to a table.
- **UPDATE**: Modifies existing data in a table.
- **DELETE**: Removes rows from a table.

**Data Control Language (DCL)**
- **GRANT**: Gives a user access privileges to the database.
- **REVOKE**: Takes back permissions granted to a user.

**Transaction Control Language (TCL)**
- **COMMIT**: Saves the changes made during a transaction.
- **ROLLBACK**: Reverts the changes made during a transaction.
- **SAVEPOINT**: Sets a point within a transaction to which you can later roll back.
- **SET TRANSACTION**: Specifies characteristics for the current transaction.

**Other Keywords**
- **WHERE**: Filters records to meet a specified condition.
- **FROM**: Indicates the table from which to retrieve or manipulate data.
- **JOIN**: Combines rows from two or more tables based on a related column.
- **UNION**: Combines the result sets of two or more SELECT statements.
- **ORDER BY**: Sorts the result set of a query.
- **GROUP BY**: Groups rows sharing a property so aggregate functions can be applied.
- **HAVING**: Filters groups formed by the GROUP BY clause.
- **DISTINCT**: Ensures that all the values are unique.
- **LIMIT**: Specifies the number of records to return.

## SQL Functions
**Aggregate Functions**
- **COUNT()**: Returns the number of rows.
- **SUM()**: Returns the total sum of a numeric column.
- **AVG()**: Returns the average value of a numeric column.
- **MIN()**: Returns the smallest value.
- **MAX()**: Returns the largest value.

**String Functions**
- **CONCAT()**: Joins two or more strings.
- **LENGTH()**: Returns the length of a string.
- **LOWER()**: Converts a string to lowercase.
- **UPPER()**: Converts a string to uppercase.
- **SUBSTRING()**: Extracts a substring from a string.
- **TRIM()**: Removes leading and trailing spaces from a string.

**Date Functions**
- **NOW()**: Returns the current date and time.
- **CURDATE()**: Returns the current date.
- **CURTIME()**: Returns the current time.
- **DATEDIFF()**: Returns the difference between two dates.
- **DATE_ADD()**: Adds a specified time interval to a date.
- **DATE_SUB()**: Subtracts a specified time interval from a date.

**Numeric Functions**
- **ABS()**: Returns the absolute value.
- **CEIL()**: Returns the smallest integer greater than or equal to a value.
- **FLOOR()**: Returns the largest integer less than or equal to a value.
- **ROUND()**: Rounds a value to the nearest integer or specified decimal places.
- **MOD()**: Returns the remainder of a division operation.
